### Has confirmed the type of the field 'date' is not correct, so we need to convert it from Float to Datetime to eliminate the (.0 after the year number: e.g 2023.0)


```python
vdg['Year'] = pd.to_datetime(vdg['Year'], format='%Y')
```

```python
vdg.head(5)
```

## Creating an barplot view of Sales by Platform for each region Japan, Europe and North America

```python
plt.rcParams['figure.figsize']=16,6
px.histogram(data_frame=vdg, x='Platform', y='JP_Sales',color='Genre', title='Sales by Platform in Japan Region')
```

We can see for Japan region, that in the top 3 platforms (DS,PS2,PS) a higher % are Role playing games. We will confirm that later on the EDA.

```python
plt.rcParams['figure.figsize']=16,6
px.histogram(data_frame=vdg, x='Platform', y='EU_Sales',color='Genre', title='Sales by Platform in Europe Region')
```

```python
plt.rcParams['figure.figsize']=16,6
px.histogram(data_frame=vdg, x='Platform', y='NA_Sales',color='Genre', title='Sales by Platform in North America Region')
```

## Has we can see comparing the 3 histograms, it appears clearly that each region due to his culture and taste of games have different top 5 platforms.
- For Japan: DS, Ps, PS2, SNES, NES (The Japanase tend to prefer Nintindo and Sony platforms)
- For Europe: PS3, PS2, XBOX360, Wii
- For North America: XBOX360, PS2, Wii, DS
